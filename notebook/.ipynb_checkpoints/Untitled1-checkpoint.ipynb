```python
def fab(n):
    a, b = 0, 1
    while n > 0:
        a, b = b, a + b
        n -= 1
    return a

print fab(0), fab(1), fab(2), fab(3), fab(4), fab(5)
import time
start_time = time.time()
print fab(10000)
end_time = time.time()
print end_time-start_time
```

```python
def cache(func):
    cached = {}
    def _func(args):
        if args not in cached:
            cached[args] = func(args)
        return cached[args]
    
    return _func

@cache
def fab(n):
    if n==0 or n==1:
        return n
    else:
        return fab(n-1) + fab(n-2)

print fab(0), fab(1), fab(2), fab(3), fab(4), fab(5)
import time
start_time = time.time()
print fab(10000)
end_time = time.time()
print end_time-start_time
```

```python
import numpy

def fab(n):
    return (numpy.matrix([[1, 1], [1, 0]]) ** (n - 1) * numpy.matrix([[1], [0]]))[0, 0]

print fab(0), fab(1), fab(2), fab(3), fab(4), fab(5)
import time
start_time = time.time()
print fab(10000)
end_time = time.time()
print end_time-start_time
```
